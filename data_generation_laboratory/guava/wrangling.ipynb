```python
# pull requests are linked to issues using syntax like "Closes #123", "Fix #123", 
# even in the github ecosystem. Here, we are just replicating that.
# please read this for more detail:
# https://docs.github.com/en/issues/tracking-your-work-with-issues/using-issues/linking-a-pull-request-to-an-issue#linking-a-pull-request-to-an-issue-using-a-keyword
```

```python
# try to find the issue number using prs
```

```python
import re
import numpy as np

def extract_issues(text):
    """
    Extracts all issue numbers from GitHub closing patterns.
    Returns a list of integers representing issue numbers.
    """
    if pd.isna(text): return np.nan
    pattern = r'''(?i)(?:close[sd]?|fixe[sd]?|resolve[sd]?)\s*[:]?\s*(?:\#(\d+)(?:\s*,\s*\#(\d+))*)'''
    
    issue_numbers = []
    
    for match in re.finditer(pattern, text, re.VERBOSE):
        # Get all numbered groups that actually matched
        groups = [g for g in match.groups() if g is not None]
        issue_numbers.extend(int(num) for num in groups)
    
    return issue_numbers if issue_numbers else np.nan
```

```python
prs['Linked_Issues'] = prs.apply(lambda x: extract_issues( x["PR_Body"] ) , axis = 1)
```

```python
prs.head()
```

```python
prs = prs[prs['Linked_Issues'].notna()]
```

```python
prs = prs.explode('Linked_Issues')
```

```python
prs["Linked_Issues"] = prs["Linked_Issues"].astype(int)
prs.head()
```

```python
issues = issues[issues['Issue_Number'].notna()].copy()

issues['Issue_Number'] = issues['Issue_Number'].astype(int)

issues.head()
```

```python
prs_issues = pd.merge(
    prs,
    issues,
    left_on='Linked_Issues',
    right_on='Issue_Number',
    how='left',
)
```
